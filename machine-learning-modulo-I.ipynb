### Comparando diferentes modelos em um único código

```
# Importar biblioteca pandas para extração dos dados
import pandas as pd

# Configuração para que nos apresente todas as 21 colunas no Display do Notebook
pd.set_option('display.max_columns', 21)

df = pd.read_csv('/content/kc_house_data.csv')
df.head()
```

```
# Separando nossas variáveis X e y

X = df.drop(['id', 'date', 'price', 'zipcode', 'lat', 'long'], axis=1)
y = df['price']

# Importar biblioteca para separar os dados de treino e teste
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=14)
```

```
# Criar função para pegar os dados de todos os modelos
def modelosRegressaoLinear(X_treino, X_teste, y_treino, y_teste):

  # Importar bibliotecas dos modelos
  from sklearn.linear_model import LinearRegression
  from sklearn.linear_model import Ridge
  from sklearn.linear_model import Lasso
  from sklearn.linear_model import ElasticNet

  # Atribuir os dados de treino e teste
  X_train = X_treino
  X_test = X_teste
  y_train = y_treino
  y_test = y_teste

  # Instaciar os modelos de regressão
  linear_regression = LinearRegression()
  ridge = Ridge()
  lasso = Lasso()
  elastic = ElasticNet()

  # Treinar os modelos
  linear_regression.fit(X_train, y_train)
  ridge.fit(X_train, y_train)
  lasso.fit(X_train, y_train)
  elastic.fit(X_train, y_train)

  # Testar os resultados
  result_linear_regression = linear_regression.score(X_test, y_test)
  result_ridge = ridge.score(X_test, y_test)
  result_lasso = lasso.score(X_test, y_test)
  result_elastic = elastic.score(X_test, y_test)

  # Modelo com maior Score
  lista_modelos = [result_linear_regression, result_ridge, result_lasso, result_elastic]
  resultado_modelo = max(lista_modelos)
  nome_modelo = []

  if result_linear_regression == resultado_modelo:
    nome_modelo.append('Regressão Linear')
  elif resultado_modelo == result_ridge:
    nome_modelo.append('Regressão Ridge')
  elif resultado_modelo == result_lasso:
    nome_modelo.append('Regressão Lasso')
  else:
    nome_modelo.append('Regressão ElasticNet')

  nome_modelo = ''.join(nome_modelo)        

  # Printar resultado dos modelos
  print(f"Resultado para o Modelo de Regressão Linear: \t {result_linear_regression} \nResultado para o Modelo de Regressão Ridge: \t {result_ridge} \nResultado para o Modelo de Regressão Lasso: \t {result_lasso} \nResultado para o Modelo de Regressão ElasticNet: {result_elastic} \n {'--'*50} \n \nO Modelo de Regressão com maior assertividade é o: {nome_modelo} com {resultado_modelo}")
```

```
modelosRegressaoLinear(X_train, X_test, y_train, y_test)
```

### Exercício de Validação Cruzada

```
# Importar biblioteca pandas para extração dos dados
import pandas as pd

# Configuração para que nos apresente todas as 21 colunas no Display do Notebook
#pd.set_option('display.max_columns', 21)

df = pd.read_csv('/content/Admission_Predict.csv')
print(df.shape)
df.head()
```

```
# Verificar valores nulos
df.isnull().sum()
```

```
# Verificar tipo dos dados
df.info()
```

```
# Separar os dados de X e y
X = df.drop(['Serial No.', 'Chance of Admit '], axis=1)
y = df['Chance of Admit '] # Variável target
```

```
# Criar função para análise de Validação Cruzada
def modelosRegressao(X, y):

  # Importar bibliotecas necessárias para análise
  from sklearn.model_selection import cross_val_score
  from sklearn.model_selection import KFold
  from sklearn.linear_model import LinearRegression
  from sklearn.linear_model import Lasso
  from sklearn.linear_model import Ridge
  from sklearn.linear_model import ElasticNet

  # Instanciar nossos modelos em variáveis
  modelo_linearR = LinearRegression()
  modelo_lasso = Lasso()
  modelo_ridge = Ridge()
  modelo_elastic = ElasticNet()

  # Aplicar KFold e definir parâmetro
  #splits_Kfold = qtd_splits_Kfold
  kfold = KFold(n_splits=5, shuffle=True)

  # Pegar resultado para cada modelo
  result_linearR = cross_val_score(modelo_linearR, X, y, cv=kfold)
  result_lasso = cross_val_score(modelo_lasso, X, y, cv=kfold)
  result_ridge = cross_val_score(modelo_ridge, X, y, cv=kfold)
  result_elastic = cross_val_score(modelo_elastic, X, y, cv=kfold)

  # Tomar o melhor resultado atavés de um dicionário
  dict_result = {"Regressão Linear":result_linearR.mean(), "Regressão Lasso":result_lasso.mean(), "Regressão Ridge":result_ridge.mean(), "Regressão ElasticNet":result_elastic.mean()}
  melhor_modelo = max(dict_result, key=dict_result.get)

  # Printar valor na tela
  print(f"A performance do Modelo de Regressão linear foi: {result_linearR.mean()}\nA performance do Modelo de Regressão Lasso foi: {result_lasso.mean()}\nA performance do Modelo de Regressão Ridge foi: {result_ridge.mean()}\nA performance do Modelo de Regressão ElasticNet foi: {result_elastic.mean()}\n {'--'*50}\n O modelo com melhor performance foi {melhor_modelo} com {dict_result[melhor_modelo]}.")               
```

```
modelosRegressao(X, y)
```

### Validação cruzada StratifiedKFold

> Possui função muito parecida com o Kfold, porém, é mais utilizada para modelos de classificação tendo em vista que se tenho uma base com 3% de aprovados e 97% de reprovados, num modelo de classificação pode impactar e essa função por debaixo dos panos vai garantir que desses 3% em cada quebra de `n_splits` tenha um mínimo de aprovados.

### Medição de Desempenho ROC/AUC

```
# Importar bibliotecas que serão utilizadas
from sklearn.datasets import load_breast_cancer
import pandas as pd

# Configuração do display
pd.set_option("display.max_columns",30)

# Instaciando dataset
dados = load_breast_cancer()

# Dividir variáveis X e y
X = pd.DataFrame(dados.data, columns=[dados.feature_names])
y = pd.Series(dados.target)
```

```
X.head()
```

```
# Importando bibliotecas
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split

# Separando os dados entre treino e teste:
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=9)

# Instanciando o modelo:
model = LogisticRegression(C=95, penalty='l1', solver='liblinear')
model.fit(X_train, y_train)

# Score:
result = model.score(X_test, y_test)
print("Acurácia:", result)
```

```
predicoes = model.predict_proba(X_test)
predicoes
```

```
probs = predicoes[:, 1]
```

```
# Importar ROC
from sklearn.metrics import roc_curve

fpr, tpr, thresholds = roc_curve(y_test, probs)
#false positive rate e true positive rate

#TPR > melhor
#FPR < melhor
```

```
print('TRP:', tpr)
print('FPR', fpr)
print('Thresholds', thresholds)
```

```
# Importar biblioteca de visualização
import matplotlib.pyplot as plt
plt.scatter(fpr,tpr)
plt.show()
```

```
from sklearn.metrics import roc_auc_score
print(roc_auc_score(y_test, probs))
```
